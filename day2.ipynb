```python
from io import StringIO
points = 0
enemy_moves = {
    'Rock': 'A',
    'Paper': 'B',
    'Scissors': 'C',
}
player_moves = {
    'Rock': 'X',
    'Paper': 'Y',
    'Scissors':'Z' 
}
move_score = {
    'Rock': 1,
    'Paper': 2,
    'Scissors': 3
}
match_score = {
    'loss': 0,
    'draw': 3,
    'win': 6
}

player_outcomes = {
    'X': 'loss',
    'Y': 'draw',
    'Z': 'win',

}
```

```python

class RockPaperScissors:
    def __init__(self, matchlist):
        self.enemy_moves = enemy_moves
        self.player_moves = player_moves
        self.move_score = move_score
        self.match_score = match_score
        self.points = 0
        self.matches = list(matchlist)
        self.enemy_turn = 0
        self.player_turn = 1
        self.enemy_move = None
        self.player_move = None
        self.match = None
        self.clean()

    def clean(self):
        # Remove all whitespace and linebreaks
        self.matches = self.matches[0:-1:2]

    def play(self):
        while self.player_turn <= len(self.matches):
            self.match = self.matches[self.enemy_turn] + self.matches[self.player_turn]
            self.get_winner()
            self.player_turn += 2
            self.enemy_turn += 2
        # print the answer (11906)
        print(self.points)
    
    def rig(self):
        while self.player_turn <= len(self.matches):
            self.match = self.matches[self.enemy_turn] + self.matches[self.player_turn]
            self.get_outcome()
            self.player_turn += 2
            self.enemy_turn += 2
        # print the answer (1186)
        print(self.points)

    def get_winner(self):
        match self.match:
            case 'AX':
                self.points += self.match_score['draw']
                self.points += self.move_score['Rock']
            case 'BX':
                self.points += self.match_score['loss']
                self.points += self.move_score['Rock']
            case 'CX':
                self.points += self.match_score['win']
                self.points += self.move_score['Rock']
            case 'AY':
                self.points += self.match_score['win']
                self.points += self.move_score['Paper']
            case 'BY':
                self.points += self.match_score['draw']
                self.points += self.move_score['Paper']
            case 'CY':
                self.points += self.match_score['loss']
                self.points += self.move_score['Paper']
            case 'AZ':
                self.points += self.match_score['loss']
                self.points += self.move_score['Scissors']
            case 'BZ':
                self.points += self.match_score['win']
                self.points += self.move_score['Scissors']
            case 'CZ':
                self.points += self.match_score['draw']
                self.points += self.move_score['Scissors']

    def get_outcome(self):
        match self.match:
            case 'AX':
                #lose against rock
                self.points += self.match_score['loss']
                self.points += self.move_score['Scissors']
            case 'BX':
                # lose against paper
                self.points += self.match_score['loss']
                self.points += self.move_score['Rock']
            case 'CX':
                # lose against scissors
                self.points += self.match_score['loss']
                self.points += self.move_score['Paper']
            case 'AY':
                # draw against rock
                self.points += self.match_score['draw']
                self.points += self.move_score['Rock']
            case 'BY':
                #draw against paper
                self.points += self.match_score['draw']
                self.points += self.move_score['Paper']
            case 'CY':
                #draw against scissors
                self.points += self.match_score['draw']
                self.points += self.move_score['Scissors']
            case 'AZ':
                # win against rock
                self.points += self.match_score['win']
                self.points += self.move_score['Paper']
            case 'BZ':
                # wins against paper
                self.points += self.match_score['win']
                self.points += self.move_score['Scissors']
            case 'CZ':
                # win against scissors
                self.points += self.match_score['win']
                self.points += self.move_score['Rock']

# get the input
with open("input2.txt", "r") as f:
    matchlist = f.read()

# initialize the game
game = RockPaperScissors(matchlist)

# run the gameloop
game.play()

# run the rigged gameloop
rigged_game = RockPaperScissors(matchlist)
rigged_game.rig()

```
